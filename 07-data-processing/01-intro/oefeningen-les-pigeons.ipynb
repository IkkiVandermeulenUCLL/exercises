```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#Jupyter notebook
```

```python
#Sample data
data = {
    'year': ['2020', '2021', '2022', '2023', '2024'],
    'Pigeons': [33, 52, 26, 40, 38],
    'Sparrows': [54, 48, 66, 59, 61],
    'Others': [104, 98, 88, 92, 101]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.plot(df["year"],df["Pigeons"], label="Pigeons")
plt.title('pigeons')
plt.legend()
plt.show()
```

```python
sparrow_plt = plt.figure()
plt.plot(df["year"], df["Sparrows"], label = "Sparrows", linestyle= "dotted", color="Green", marker= ">", markeredgecolor = "red")
plt.title("Sparrows")
plt.show()
```

```python
barplot = plt.subplot()
barplot.bar(df['year'],df['Pigeons'], color="Blue")
barplot.bar(df['year'],df['Sparrows'], color="Purple")
barplot.bar(df['year'],df['Others'], color="Green")
plt.show()
```
